```python
import sqlite3
import pandas as pd


```

```python
# Load the school ranking data from the CSV file into a DataFrame
df = pd.read_csv('school_ranking_cleaned.csv')
```

Generate a list of tuples from the dataframe

```python
data_tuples = []

for i, row in enumerate(df.itertuples(), 1):
    data_tuple = (
        i,
        row.school_name,
        row.rank_detail_url,
        row.school_rating,
        row.school_rank,
        row.city,
        row.school_group,
        row.school_type,
        row.school_id
        )
    data_tuples.append(data_tuple)
```

```python
len(data_tuples)
```

```python
data_tuples
```

```python
conn = sqlite3.connect('../database.db')
```

```python
cur = conn.cursor()
```

```python
res = cur.execute('DROP TABLE school_ranking')
```

```python
cur.execute("""
            CREATE TABLE IF NOT EXISTS school_ranking (
            id INTEGER PRIMARY KEY,
            school_name TEXT,
            rank_detail_url TEXT,
            school_rating REAL,
            school_rank TEXT,
            city TEXT,
            school_group TEXT,
            school_type TEXT,
            school_id INTEGER,
            FOREIGN KEY(school_id) REFERENCES schools(school_id)
            )
            """)
```

```python
# use executemany to insert the data into the database
cur.executemany('''
    INSERT INTO school_ranking VALUES (?,?,?,?,?,?,?,?,?)
''', data_tuples)
```

```python
conn.commit()
```

```python
conn.close()
```
